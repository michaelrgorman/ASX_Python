```python
#https://www.youtube.com/watch?v=t_vZDyQDUkk
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
```

```python
end = dt.datetime.now()
start = dt.datetime(2000,1,1)
start, end
```

```python
stockList = ['CBA', 'NAB', 'ANZ', 'BOQ', 'BEN', 'WBC']
stocks = [i + '.AX' for i in stockList]
stocks
```

```python
df = pdr.get_data_yahoo(stocks, start, end)
df.head()
```

```python
df.index
```

```python
df.columns
```

```python
Close = df.Close
Close.head()
```

```python
Close.describe()
```

```python
Close[Close.index > end - dt.timedelta(days = 100)].describe()
```

```python
Close.plot(figsize=(12,8))
```
